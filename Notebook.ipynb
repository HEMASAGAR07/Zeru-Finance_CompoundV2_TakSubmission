```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import json
import datetime as dt
from google.colab import drive, files
import warnings
warnings.filterwarnings('ignore')

# Function to load and parse JSON data
def load_json_data(file_path):
    print(f"Loading data from {file_path}...")
    with open(file_path, 'r') as f:
        data = json.load(f)
    return data

# Load the three largest files from the dataset
file_paths = [
    '/content/Compound_V2_Raw_Dataset/compoundV2_transactions_ethereum_chunk_0.json',
    '/content/Compound_V2_Raw_Dataset/compoundV2_transactions_ethereum_chunk_1.json',
    '/content/Compound_V2_Raw_Dataset/compoundV2_transactions_ethereum_chunk_2.json'
]

# Process transaction data from JSON files
def process_transactions(json_data):
    transactions = []

    # Process deposits
    if 'deposits' in json_data:
        for deposit in json_data['deposits']:
            transactions.append({
                'wallet': deposit['account']['id'],
                'action': 'deposit',
                'value': float(deposit['amount']),
                'value_usd': float(deposit['amountUSD']),
                'asset': deposit['asset']['symbol'],
                'timestamp': int(deposit['timestamp']),
                'hash': deposit['hash']
            })

    # Process borrows
    if 'borrows' in json_data:
        for borrow in json_data['borrows']:
            transactions.append({
                'wallet': borrow['account']['id'],
                'action': 'borrow',
                'value': float(borrow['amount']),
                'value_usd': float(borrow['amountUSD']),
                'asset': borrow['asset']['symbol'],
                'timestamp': int(borrow['timestamp']),
                'hash': borrow['hash']
            })

    # Process repays
    if 'repays' in json_data:
        for repay in json_data['repays']:
            transactions.append({
                'wallet': repay['account']['id'],
                'action': 'repay',
                'value': float(repay['amount']),
                'value_usd': float(repay['amountUSD']),
                'asset': repay['asset']['symbol'],
                'timestamp': int(repay['timestamp']),
                'hash': repay['hash']
            })

    # Process withdraws
    if 'withdraws' in json_data:
        for withdraw in json_data['withdraws']:
            transactions.append({
                'wallet': withdraw['account']['id'],
                'action': 'withdraw',
                'value': float(withdraw['amount']),
                'value_usd': float(withdraw['amountUSD']),
                'asset': withdraw['asset']['symbol'],
                'timestamp': int(withdraw['timestamp']),
                'hash': withdraw['hash']
            })

    # Process liquidations
    if 'liquidations' in json_data:
        for liquidation in json_data['liquidations']:
            transactions.append({
                'wallet': liquidation['borrower']['id'],
                'action': 'liquidation',
                'value': float(liquidation['amount']),
                'value_usd': float(liquidation['amountUSD']),
                'asset': liquidation['asset']['symbol'],
                'timestamp': int(liquidation['timestamp']),
                'hash': liquidation['hash']
            })

    return transactions

# Load and process all files
all_transactions = []
for file_path in file_paths:
    try:
        json_data = load_json_data(file_path)
        transactions = process_transactions(json_data)
        all_transactions.extend(transactions)
        print(f"Processed {len(transactions)} transactions from {file_path}")
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

# Convert to DataFrame
df = pd.DataFrame(all_transactions)

# Convert timestamp to datetime
df['datetime'] = pd.to_datetime(df['timestamp'], unit='s')
print(f"Total transactions loaded: {len(df)}")

# Display sample data
print("\nSample data:")
display(df.head())

# Feature Engineering
def extract_wallet_features(df):
    print("Extracting wallet-level features...")
    wallet_features = {}

    # Get unique wallets
    wallets = df['wallet'].unique()

    for wallet in wallets:
        wallet_data = df[df['wallet'] == wallet]

        # Skip wallets with too few transactions
        if len(wallet_data) < 3:
            continue

        # 1. Transaction Pattern Features
        actions = wallet_data['action'].value_counts().to_dict()
        deposit_count = actions.get('deposit', 0)
        withdraw_count = actions.get('withdraw', 0)
        borrow_count = actions.get('borrow', 0)
        repay_count = actions.get('repay', 0)
        liquidation_count = actions.get('liquidation', 0)

        # Calculate transaction ratios
        deposit_withdraw_ratio = deposit_count / max(withdraw_count, 1)
        borrow_repay_ratio = borrow_count / max(repay_count, 1)

        # Transaction value statistics
        value_stats = wallet_data['value_usd'].describe()

        # 2. Risk Behavior Features
        # Liquidation frequency
        liquidation_frequency = liquidation_count / len(wallet_data)

        # 3. Temporal Stability Features
        # Account age
        time_range = (wallet_data['datetime'].max() - wallet_data['datetime'].min()).days
        account_age = max(time_range, 1)  # Avoid division by zero

        # Transaction frequency
        transactions_per_day = len(wallet_data) / account_age

        # Transaction regularity (standard deviation of time between transactions)
        wallet_data = wallet_data.sort_values('timestamp')
        if len(wallet_data) > 1:
            time_diffs = np.diff(wallet_data['timestamp'])
            time_regularity = np.std(time_diffs) / (np.mean(time_diffs) + 1)  # Normalized std
        else:
            time_regularity = 0

        # 4. Protocol Interaction Features
        # Asset diversity
        asset_diversity = wallet_data['asset'].nunique()

        # Calculate position sizes relative to account average
        position_volatility = wallet_data.groupby('action')['value_usd'].std().mean() / (wallet_data['value_usd'].mean() + 1)

        # Store features for this wallet
        wallet_features[wallet] = {
            'transaction_count': len(wallet_data),
            'deposit_count': deposit_count,
            'withdraw_count': withdraw_count,
            'borrow_count': borrow_count,
            'repay_count': repay_count,
            'liquidation_count': liquidation_count,
            'deposit_withdraw_ratio': deposit_withdraw_ratio,
            'borrow_repay_ratio': borrow_repay_ratio,
            'mean_transaction_value': value_stats['mean'],
            'max_transaction_value': value_stats['max'],
            'std_transaction_value': value_stats['std'],
            'liquidation_frequency': liquidation_frequency,
            'account_age_days': account_age,
            'transactions_per_day': transactions_per_day,
            'time_regularity': time_regularity,
            'asset_diversity': asset_diversity,
            'position_volatility': position_volatility
        }

    # Convert to DataFrame
    wallet_features_df = pd.DataFrame.from_dict(wallet_features, orient='index')

    # Derive additional features
    wallet_features_df['repayment_consistency_ratio'] = wallet_features_df['repay_count'] / wallet_features_df['borrow_count'].replace(0, 1)
    wallet_features_df['collateral_health_score'] = 1 / (wallet_features_df['liquidation_frequency'] + 0.01)
    wallet_features_df['behavioral_stability_score'] = (
        wallet_features_df['account_age_days'] *
        wallet_features_df['transactions_per_day'] /
        (wallet_features_df['position_volatility'] + 1)
    )

    # Cap extreme values
    for col in wallet_features_df.columns:
        if wallet_features_df[col].dtype in [np.float64, np.int64]:
            q1 = wallet_features_df[col].quantile(0.01)
            q3 = wallet_features_df[col].quantile(0.99)
            wallet_features_df[col] = wallet_features_df[col].clip(q1, q3)

    return wallet_features_df

# Extract features
wallet_features_df = extract_wallet_features(df)
print(f"Extracted features for {len(wallet_features_df)} wallets")

# Display feature statistics
print("\nFeature statistics:")
display(wallet_features_df.describe())

# Prepare data for clustering
def prepare_for_clustering(df):
    print("Preparing data for clustering...")
    # Select numerical features
    numerical_features = df.select_dtypes(include=[np.number])

    # Fill any NaN values with 0
    numerical_features = numerical_features.fillna(0)

    # Standardize features
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(numerical_features)

    # Apply PCA for visualization
    pca = PCA(n_components=2)
    principal_components = pca.fit_transform(scaled_features)

    return scaled_features, principal_components, scaler

# Prepare data
scaled_features, principal_components, scaler = prepare_for_clustering(wallet_features_df)

# Determine optimal number of clusters using the elbow method
def find_optimal_clusters(scaled_features, max_clusters=10):
    print("Finding optimal number of clusters...")
    inertia = []
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(scaled_features)
        inertia.append(kmeans.inertia_)

    # Plot elbow curve
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, max_clusters + 1), inertia, marker='o')
    plt.title('Elbow Method for Optimal k')
    plt.xlabel('Number of clusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()

    # Calculate the rate of decrease
    inertia_diff = np.diff(inertia)
    inertia_diff2 = np.diff(inertia_diff)

    # Find the elbow point (where the rate of decrease significantly changes)
    optimal_k = np.argmax(inertia_diff2) + 2
    return min(optimal_k, 5)  # Cap at 5 clusters for interpretability

# Find optimal clusters
optimal_k = find_optimal_clusters(scaled_features)
print(f"Optimal number of clusters: {optimal_k}")

# Perform clustering
def perform_clustering(scaled_features, n_clusters):
    print(f"Performing clustering with {n_clusters} clusters...")
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(scaled_features)

    # Visualize clusters
    plt.figure(figsize=(12, 8))
    scatter = plt.scatter(principal_components[:, 0], principal_components[:, 1],
                          c=cluster_labels, cmap='viridis', alpha=0.7)
    plt.title(f'Wallet Clusters (k={n_clusters})')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.colorbar(scatter, label='Cluster')
    plt.grid(True, alpha=0.3)
    plt.show()

    return cluster_labels, kmeans

# Perform clustering
cluster_labels, kmeans_model = perform_clustering(scaled_features, optimal_k)

# Add cluster labels to features
wallet_features_df['cluster'] = cluster_labels

# Analyze clusters
def analyze_clusters(df):
    print("Analyzing clusters...")
    cluster_stats = df.groupby('cluster').mean()

    # Display key metrics by cluster
    plt.figure(figsize=(15, 10))

    # Plot liquidation frequency by cluster
    plt.subplot(2, 2, 1)
    sns.barplot(x=cluster_stats.index, y=cluster_stats['liquidation_frequency'])
    plt.title('Liquidation Frequency by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Avg. Liquidation Frequency')

    # Plot repayment consistency by cluster
    plt.subplot(2, 2, 2)
    sns.barplot(x=cluster_stats.index, y=cluster_stats['repayment_consistency_ratio'])
    plt.title('Repayment Consistency by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Avg. Repayment Consistency')

    # Plot transaction count by cluster
    plt.subplot(2, 2, 3)
    sns.barplot(x=cluster_stats.index, y=cluster_stats['transaction_count'])
    plt.title('Transaction Count by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Avg. Transaction Count')

    # Plot account age by cluster
    plt.subplot(2, 2, 4)
    sns.barplot(x=cluster_stats.index, y=cluster_stats['account_age_days'])
    plt.title('Account Age by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Avg. Account Age (days)')

    plt.tight_layout()
    plt.show()

    return cluster_stats

# Analyze clusters
cluster_stats = analyze_clusters(wallet_features_df)

# Assign base scores to clusters based on their characteristics
def assign_base_scores(cluster_stats):
    print("Assigning base scores to clusters...")
    # Normalize key metrics
    normalized_stats = cluster_stats.copy()

    # Invert metrics where lower is better
    normalized_stats['liquidation_frequency'] = 1 - normalized_stats['liquidation_frequency'] / normalized_stats['liquidation_frequency'].max()
    normalized_stats['position_volatility'] = 1 - normalized_stats['position_volatility'] / normalized_stats['position_volatility'].max()

    # Normalize metrics where higher is better
    for col in ['repayment_consistency_ratio', 'collateral_health_score', 'behavioral_stability_score', 'asset_diversity']:
        normalized_stats[col] = normalized_stats[col] / normalized_stats[col].max()

    # Calculate base score as weighted average of key metrics
    normalized_stats['base_score'] = (
        normalized_stats['liquidation_frequency'] * 0.3 +
        normalized_stats['repayment_consistency_ratio'] * 0.25 +
        normalized_stats['collateral_health_score'] * 0.2 +
        normalized_stats['behavioral_stability_score'] * 0.15 +
        normalized_stats['asset_diversity'] * 0.1
    )

    # Scale to 0-100 range
    min_score = normalized_stats['base_score'].min()
    max_score = normalized_stats['base_score'].max()
    normalized_stats['base_score'] = 20 + 60 * (normalized_stats['base_score'] - min_score) / (max_score - min_score)

    # Display base scores
    plt.figure(figsize=(10, 6))
    sns.barplot(x=normalized_stats.index, y=normalized_stats['base_score'])
    plt.title('Base Score by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Base Score (0-100)')
    plt.grid(True, alpha=0.3)
    plt.show()

    return normalized_stats['base_score'].to_dict()

# Assign base scores
cluster_base_scores = assign_base_scores(cluster_stats)

# Calculate final credit scores
def calculate_credit_scores(df, cluster_base_scores):
    print("Calculating final credit scores...")

    # Define scoring function
    def score_wallet(row):
        # Get base score from cluster
        base_score = cluster_base_scores[row['cluster']]

        # Apply multipliers
        multipliers = [
            1 - (0.2 * min(row['liquidation_count'], 5)),  # Cap at 5 liquidations
            1 + (0.1 * min(row['repayment_consistency_ratio'], 2)),  # Cap at 2x ratio
            1 + (0.15 * min(row['collateral_health_score'] / 100, 1.5)),  # Cap at 1.5x boost
            1 + (0.1 * min(row['behavioral_stability_score'], 1))  # Cap at 1x boost
        ]

        # Calculate raw score
        raw_score = base_score * np.prod(multipliers)

        # Normalize to 0-100 scale
        return max(0, min(100, raw_score))

    # Calculate scores
    df['credit_score'] = df.apply(score_wallet, axis=1)

    # Round scores to integers
    df['credit_score'] = df['credit_score'].round().astype(int)

    # Display score distribution
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.histplot(df['credit_score'], bins=20, kde=True)
    plt.title('Credit Score Distribution')
    plt.xlabel('Credit Score')
    plt.ylabel('Count')

    plt.subplot(1, 2, 2)
    sns.boxplot(x='cluster', y='credit_score', data=df)
    plt.title('Credit Score by Cluster')
    plt.xlabel('Cluster')
    plt.ylabel('Credit Score')

    plt.tight_layout()
    plt.show()

    return df

# Calculate credit scores
wallet_features_df = calculate_credit_scores(wallet_features_df, cluster_base_scores)

# Generate top 1000 wallets CSV
def generate_top_wallets_csv(df, n=1000):
    print(f"Generating CSV with top {n} wallets...")
    # Sort by credit score (descending)
    top_wallets = df.sort_values('credit_score', ascending=False).head(n)

    # Reset index to make wallet address a column
    top_wallets = top_wallets.reset_index().rename(columns={'index': 'wallet_address'})

    # Select relevant columns
    output_columns = ['wallet_address', 'credit_score', 'cluster',
                     'transaction_count', 'liquidation_count',
                     'repayment_consistency_ratio', 'collateral_health_score']

    top_wallets_output = top_wallets[output_columns]

    # Save to CSV
    output_filename = 'compound_v2_top_wallets.csv'
    top_wallets_output.to_csv(output_filename, index=False)

    # Download file in Colab
    files.download(output_filename)

    return top_wallets_output

# Generate top wallets CSV
top_wallets = generate_top_wallets_csv(wallet_features_df)

# Analyze high and low scoring wallets
def analyze_extreme_wallets(df, n=5):
    print(f"Analyzing top and bottom {n} wallets...")

    # Get top and bottom wallets
    top_wallets = df.sort_values('credit_score', ascending=False).head(n)
    bottom_wallets = df.sort_values('credit_score', ascending=True).head(n)

    # Display comparison
    comparison_metrics = ['transaction_count', 'deposit_count', 'withdraw_count',
                         'borrow_count', 'repay_count', 'liquidation_count',
                         'deposit_withdraw_ratio', 'borrow_repay_ratio',
                         'liquidation_frequency', 'account_age_days',
                         'transactions_per_day', 'repayment_consistency_ratio',
                         'collateral_health_score', 'behavioral_stability_score']

    # Create comparison table
    comparison = pd.DataFrame({
        'Top Wallets (Avg)': top_wallets[comparison_metrics].mean(),
        'Bottom Wallets (Avg)': bottom_wallets[comparison_metrics].mean(),
        'Ratio (Top/Bottom)': top_wallets[comparison_metrics].mean() / bottom_wallets[comparison_metrics].mean().replace(0, 0.001)
    })

    # Display table
    print("\nComparison of Top vs Bottom Wallets:")
    display(comparison)

    # Visualize key differences
    plt.figure(figsize=(14, 8))

    # Select metrics with largest differences
    key_metrics = comparison.sort_values('Ratio (Top/Bottom)', ascending=False).head(6).index

    for i, metric in enumerate(key_metrics):
        plt.subplot(2, 3, i+1)
        data = pd.DataFrame({
            'Top': top_wallets[metric],
            'Bottom': bottom_wallets[metric]
        }).melt()
        sns.boxplot(x='variable', y='value', data=data)
        plt.title(metric)
        plt.xlabel('')
        plt.xticks(rotation=45)

    plt.tight_layout()
    plt.show()

    return top_wallets, bottom_wallets

# Analyze extreme wallets
top_wallets_sample, bottom_wallets_sample = analyze_extreme_wallets(wallet_features_df)

# Display final score distribution
plt.figure(figsize=(10, 6))
sns.histplot(wallet_features_df['credit_score'], bins=20, kde=True)
plt.axvline(wallet_features_df['credit_score'].mean(), color='r', linestyle='--', label=f'Mean: {wallet_features_df["credit_score"].mean():.1f}')
plt.axvline(wallet_features_df['credit_score'].median(), color='g', linestyle='--', label=f'Median: {wallet_features_df["credit_score"].median():.1f}')
plt.title('Compound V2 Wallet Credit Score Distribution')
plt.xlabel('Credit Score (0-100)')
plt.ylabel('Count')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
